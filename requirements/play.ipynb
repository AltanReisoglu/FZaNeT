```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision.models import wide_resnet50_2,resnet50,ResNet50_Weights,Wide_ResNet50_2_Weights
```

```python
class MyModule(nn.Module):
    """Some Information about MyModule"""
    def __init__(self):
        super(MyModule, self).__init__()
        self.base_model=wide_resnet50_2(weights=Wide_ResNet50_2_Weights)
    def forward(self, x):
        self.base_model(x)
        return x
```
